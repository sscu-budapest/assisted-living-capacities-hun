```python
import datetime as dt
import datazimmer as dz
from tqdm.notebook import tqdm
from itertools import count

import pandas as pd
from aswan import get_soup
```

```python
url = "https://nyugdijasotthon.hu/otthon-kereses?title=&field_otthon_megye_tid=All&field_otthon_varakozasi_ido_tid=All&field_otthon_tulajdonos_tipusa_tid=All"
```

```python
dfs = [pd.read_html(url)[0]]

for i in tqdm(count(1)):
    try:
        dfs.append(pd.read_html(url + f"&page={i}")[0])
    except:
        break
```

```python
df = pd.concat(dfs).drop_duplicates()
```
